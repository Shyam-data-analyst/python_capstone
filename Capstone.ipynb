3) Exploratory Data Analysis (EDA) and Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Campaign Performance:
➢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and
ROI.
➢ Create a bar chart to visualize the average Conversion_Rate for different
Channel_Used, categorized by Campaign_Type.
➢ Visualize the distribution of Engagement_Score across different Campaign_Type
using a box plot.
➢ Analyze the average ROI by Company using a bar chart to compare the
profitability of campaigns conducted by different companies.
➢ Examine the correlation between Engagement_Score and Conversion_Rate
using a heatmap.

```python
plt.figure(figsize=(12,6))
plt.title('Acquisition_Cost vs ROI',fontsize = '10',fontweight ='bold')
sns.scatterplot(x='Acquisition_Cost',y='ROI',data = df)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Channel_Used',y='Conversion_Rate',data = df,hue='Campaign_Type',estimator='mean',palette=['r','b','g','gray','y'])
plt.title("Channel Used")
plt.ylim(0,6)
plt.show() 
```

```python
plt.figure(figsize=(12,6))
plt.title("Distribution across Campaign type",fontsize = 10,fontweight ='bold')
plt.xticks(rotation=45)
sns.boxplot(x='Campaign_Type',y='Engagement_Score',data = df,hue='Campaign_Type')
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title("Comparison of average ROI by Company and Campaigns")
sns.barplot(x='Company',y='ROI',estimator='mean',data=df,hue='Campaign_Type')
plt.show()
```

```python
Correlation = df[['Engagement_Score','Conversion_Rate']].corr()
Correlation
```

```python
sns.heatmap(Correlation,annot=True,cmap='coolwarm')
```

Customer Segmentation:
➢ Create a count plot to visualize the distribution of Target_Audience.
➢ Identify which Customer_Segment has the highest Conversion_Rate for each
Language using a bar chart.
➢ Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.
➢ Analyze average Conversion_Rate by Language using a bar chart to compare
the effectiveness of campaigns conducted in different languages.

```python
plt.figure(figsize=(12,6))
ax=sns.countplot(x='Target_Audience',data =df,width=0.5,color='Gray')
ax.bar_label(ax.containers[0])
plt.title("Target Count",fontsize=10,fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.ylim(0,9)
plt.title("Customer Segment with Highest Conversion Rate for Each Language",fontweight='bold')
ax = sns.barplot(x='Customer_Segment',y='Conversion_Rate',hue='Language',data=df,estimator='max',width=0.9)
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Distribution of Acquisition Cost across Customer Segments by Channel Used",fontweight='bold')
sns.boxplot(x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used',data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
#doubtttt
plt.figure(figsize=(16,6))
plt.title("Average Conversion Rate by Language and Campaigns",fontweight='bold')
ax=sns.barplot(x='Language',y='Conversion_Rate',data =df,hue='Campaign_Type',width = 0.8)
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f', padding=4)
plt.show()
```

Channel Effectiveness:

```python

plt.figure(figsize=(12,6))
plt.title("Comparison of Engagement Score for different Channels Used and Campaign Type",fontweight='bold')
sns.barplot(x='Channel_Used',y='Engagement_Score',hue='Campaign_Type',data =df,palette='Set2')
plt.show()
```

```python
pie = df.groupby('Channel_Used')['ROI'].sum()
plt.title("Distribution of total ROI across Channels Used")
plt.pie(pie,labels=(pie.index),autopct='%1.2f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Relationship between Clicks and Impressions by Channel Used",fontweight='bold')
sns.scatterplot(x='Clicks',y='Impressions',hue='Channel_Used',data =df)
plt.show()
```

Time-Based Analysis:
➢ Plot the distribution of Duration using a histogram.
➢ Analyze how the overall Conversion_Rate has changed over Date for each
Company using a line chart.
➢ Examine the trend of Engagement_Score over Date with a lineline

```python
plt.figure(figsize=(12,6))
plt.title("Distribution of Duration")
sns.histplot(x="Duration",data =df,kde=True,color='Green')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Analysing overall Conversion Rate of Company by Date",fontweight='bold')
sns.lineplot(x='Date',y='Conversion_Rate',data =df,hue='Company',style='Company',marker='o',errorbar=None)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Trend of Engagement_Score over Date")
sns.lineplot(x='Date',y='Engagement_Score',data =df,errorbar=None,marker='o',markeredgecolor='r',linestyle='--')
plt.show()
```

Geographic Analysis:

```python
plt.figure(figsize=(15,6))
plt.title("Locations cost",fontweight='bold')
plt.xticks(rotation=45)
ax =sns.barplot(x='Location',y='Acquisition_Cost',hue='Location',data = df,errorbar=None)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title("Visualizing the Conversion_Rate by Location and Target Audience",fontweight='bold')
plt.xticks(rotation=45)
sns.barplot(x='Location',y='Conversion_Rate',hue='Target_Audience',data =df,palette='Set3',width=0.8)
plt.show()
```

```python
proportion = df.groupby('Location')['ROI'].sum()
proportion
```

```python
plt.figure(figsize=(15,6))
plt.title("Proportion of ROI by Location",fontweight='bold')
plt.pie(proportion,labels=proportion.index,autopct='%1.2f%%',shadow=True,startangle=90)
plt.show()
```
